```python
import matplotlib.pyplot as plt
import numpy as np

from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)
```

# Data Points

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])
```

# Marker

## *marker*
- used to add points on each edge of the line
- the type of marker needed can be passed in 

```python
plt.plot(x, y, marker='o')
plt.show()
```

## *markersize*
- the size of the plotted marker can be changed

```python
plt.plot(x, y, marker='o', markersize=12)
plt.show()
```

## *markeredgecolor*
- the edge of the marker can be colored

```python
plt.plot(x, y, marker = 'o', markeredgecolor = 'r')
plt.show()
```

## *makerfacecolor*
- the inside of the maker can be filled with a color

```python
plt.plot(x, y, marker = 'o', markerfacecolor = 'r')
plt.show()
```

# Line

### plotting without a line

```python
plt.plot(x, y, 'o') 

plt.show() 
```

## setting width of the plotted line

## *linwidth*
- to increase the width of the plotted line linwidth is used

```python
plt.plot(x, y, linewidth=2.0)

plt.show()
```

## coloring plotted line

##*color*
- change the color of the plotted line


```python
plt.plot(x, y, color='r')
plt.show()
```

## changing type of plotted line

##*linestyle*
- the type of line plotted can be changed

```python
plt.plot(x, y, linestyle=':')
plt.show()
```

# Grid Lines

## *grid()*
- used to display the grid in the graph

```python
plt.plot(x, y)

plt.grid()

plt.show()
```

## *axis*
- used to specify along which axis we need the grid

```python
plt.plot(x, y)

plt.grid(axis = 'x')

plt.show()
```

## setting line properties of grid

## *color*
- coloring the grid line

## *linestyle*
- type of grid line

## *linewidth*
- width of the grid line

```python
plt.plot(x, y)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```
