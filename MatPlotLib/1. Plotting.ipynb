# Data Points

```python
x = np.array([0, 6])   # creating points for x-axis
y = np.array([0, 250]) # creatinmg points for y-axis
```

# Plotting 2 data points
## Graphing x and y points 

### plot()
- is used to draw points (markers) in a diagram
- draws a line from point to point.

- Parameter 1: array of points on x-axis
- Parameter 2: array of points on y-axis \

To draw a line from (1, 3) to (8, 10), \
we need arrays [1, 8] and [3, 10] to the plot \\ function.

### show()
- it displays the plotted graph

```python
plt.plot(x, y) 

plt.show() 
```

# Plotting multiple data points

- plotting a graph from (1,3) -> (2,8) -> (6,1) -> (8,10)

```python
x = np.array([1, 2, 6, 8])
y = np.array([3, 8, 1, 10])

plt.plot(x, y)

plt.show()
```

# Default x-points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, 

```python
y = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y)

plt.show()
```

# Plotting Multiple Lines

```python
x = np.array([3, 8, 1, 10])
y = np.array([6, 2, 7, 11])

plt.plot(x)
plt.plot(y)

plt.show()
```

```python
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```

# Display Multiple Plots

## *subplot()*

- allows you to draw multiple plots in one figure
- Parameter 1: #rows in the plot
- Parameter 2: #columns in the plot
- Parameter 3: plot number

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```
