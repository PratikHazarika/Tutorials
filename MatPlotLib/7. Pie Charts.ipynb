```python
import matplotlib.pyplot as plt
import numpy as np

from google.colab import drive
drive.mount('/content/gdrive/', force_remount=True)
```

# Creating Pie Chart

## *pie*
- used to create a pie chart

```python
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

# Labeling Pie Chart

## *labels*
- labels can be passed for each of the sections

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

# Coloring Sections

## *colors* 
- each section of the pie chart can be given colors

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["black", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = mylabels, colors = mycolors)
plt.show() 
```

# Starting Angle

## *startangle*
- by default, the start angle is x-axis i.e. 0
- the starting angle can be change w.r.t x-axis

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show() 
```

# Separating Sections

## *explode*
- used to make sections of the pie chart be separted 
- the separation distance from the centre must be passsed

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels=mylabels, explode=myexplode)
plt.show() 
```

# Shadow

## *shadow*
- shadows can be added to separted sections of the chart

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show() 
```

# Legend

## *legend()*
- legend of the colors can be added to the chart

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.legend(title = "Four Fruits:")
plt.show() 
```
